```python
import numpy as np

def convert_mpb_to_physical_tensor(k_mpb, f_mpb, E_mpb, H_mpb, a, grid_spacing, epsilon, mu_r=1.0):
    """
    Convert normalized MPB quantities to physical units with tensor permittivity.

    Parameters
    ----------
    k_mpb : float or np.ndarray
        Normalized wavevector from MPB (dimensionless).
        
    f_mpb : float or np.ndarray
        Normalized frequency from MPB (dimensionless).
        
    E_mpb : np.ndarray
        Normalized electric field from MPB (dimensionless).
        Expected shape: (Nx, Ny, Nz, 3)
        
    H_mpb : np.ndarray
        Normalized magnetic field from MPB (dimensionless).
        Expected shape: (Nx, Ny, Nz, 3)
        
    a : float
        Lattice constant of the photonic crystal (meters).
        
    grid_spacing : tuple of floats
        The grid spacing in each spatial dimension (dx, dy, dz) in meters.
        
    epsilon : np.ndarray
        Permittivity tensor field.
        Expected shape: (Nx, Ny, Nz, 3, 3)
        If scalar permittivity, it can be provided as a scalar or broadcastable array.
        
    mu_r : float, optional
        Relative permeability of the medium. Default is 1.0 (vacuum).
        
    Returns
    -------
    k_phys : float or np.ndarray
        Physical wavevector in radians per meter (rad/m).
        
    f_phys : float or np.ndarray
        Physical frequency in Hertz (Hz).
        
    E_phys : np.ndarray
        Physical electric field in Volts per meter (V/m).
        Shape: (Nx, Ny, Nz, 3)
        
    H_phys : np.ndarray
        Physical magnetic field in Amperes per meter (A/m).
        Shape: (Nx, Ny, Nz, 3)
    """
    
    # Physical Constants
    c = 2.99792458e8          # Speed of light in vacuum (m/s)
    epsilon0 = 8.854187817e-12 # Vacuum permittivity (F/m)
    mu0 = 4 * np.pi * 1e-7     # Vacuum permeability (H/m)
    
    # Calculate medium properties
    mu = mu0 * mu_r
    
    # Convert normalized wavevector to physical wavevector
    # k_phys = k_mpb * (2 * pi / a) radians per meter
    k_phys = k_mpb * (2 * np.pi) / a
    
    # Convert normalized frequency to physical frequency
    # f_phys = f_mpb * (c / a) Hertz
    f_phys = f_mpb * c / a
    
    # Calculate voxel volume
    dx, dy, dz = grid_spacing
    voxel_volume = dx * dy * dz  # m^3
    
    # Ensure E_mpb and H_mpb are numpy arrays
    E_mpb = np.array(E_mpb)
    H_mpb = np.array(H_mpb)
    
    # Ensure epsilon is a numpy array
    epsilon = np.array(epsilon)
    
    # Handle scalar permittivity by expanding dimensions
    if epsilon.ndim == 0:
        epsilon = epsilon * np.ones(E_mpb.shape + (3,3))
    elif epsilon.shape == E_mpb.shape + (3,3):
        pass  # Correct shape
    else:
        raise ValueError(f"Permittivity tensor shape {epsilon.shape} is incompatible with E_mpb shape {E_mpb.shape}")
    
    # Compute (E_mpb)^T @ epsilon @ E_mpb for each grid point
    # Using einsum for efficient computation
    # 'ijkl,ijklm,ijklm->ijk' where:
    # i,j,k are spatial indices
    # l is vector component for E
    # m is tensor component for epsilon
    E_contracted = np.einsum('ijkl,ijklm,ijklm->ijk', E_mpb, epsilon, E_mpb)
    
    # Compute sum_epsilon_E2 * voxel_volume
    sum_epsilon_E2 = np.sum(E_contracted) * voxel_volume
    
    # Compute scaling factor for E
    scaling_E = 1.0 / np.sqrt(sum_epsilon_E2)
    
    # Scale E_mpb
    E_phys = E_mpb * scaling_E
    
    # Compute sum_H2 * voxel_volume
    sum_H2 = np.sum(np.abs(H_mpb)**2) * voxel_volume
    
    # Compute scaling factor for H
    scaling_H = 1.0 / np.sqrt(sum_H2)
    
    # Scale H_mpb
    H_phys = H_mpb * scaling_H
    
    return k_phys, f_phys, E_phys, H_phys
```

```python
import numpy as np

def convert_mpb_to_physical_2d(k_mpb, f_mpb, E_mpb, H_mpb, a, grid_spacing, epsilon, mu_r=1.0, mode='TE'):
    """
    Convert normalized MPB quantities to physical units for 2D photonic crystals.

    Parameters
    ----------
    k_mpb : float or np.ndarray
        Normalized wavevector from MPB (dimensionless).
        
    f_mpb : float or np.ndarray
        Normalized frequency from MPB (dimensionless).
        
    E_mpb : np.ndarray
        Normalized electric field from MPB (dimensionless).
        - TE Mode: Shape (Nx, Ny, 2) for E_x, E_y
        - TM Mode: Shape (Nx, Ny, 1) for E_z
        
    H_mpb : np.ndarray
        Normalized magnetic field from MPB (dimensionless).
        - TE Mode: Shape (Nx, Ny, 1) for H_z
        - TM Mode: Shape (Nx, Ny, 2) for H_x, H_y
        
    a : float
        Lattice constant of the photonic crystal (meters).
        
    grid_spacing : tuple of floats
        The grid spacing in each spatial dimension (dx, dy) in meters.
        
    epsilon : np.ndarray
        Permittivity tensor field.
        - TE Mode: Shape (Nx, Ny, 2, 2)
        - TM Mode: Shape (Nx, Ny, 1, 1) or expanded for anisotropy
        
    mu_r : float, optional
        Relative permeability of the medium. Default is 1.0 (vacuum).
        
    mode : str, optional
        Polarization mode: 'TE' or 'TM'. Default is 'TE'.
        
    Returns
    -------
    k_phys : float or np.ndarray
        Physical wavevector in radians per meter (rad/m).
        
    f_phys : float or np.ndarray
        Physical frequency in Hertz (Hz).
        
    E_phys : np.ndarray
        Physical electric field in Volts per meter (V/m).
        
    H_phys : np.ndarray
        Physical magnetic field in Amperes per meter (A/m).
    """
    
    # Validate mode
    if mode not in ['TE', 'TM']:
        raise ValueError("Mode must be either 'TE' or 'TM'")
    
    # Physical Constants
    c = 2.99792458e8          # Speed of light in vacuum (m/s)
    epsilon0 = 8.854187817e-12 # Vacuum permittivity (F/m)
    mu0 = 4 * np.pi * 1e-7     # Vacuum permeability (H/m)
    
    # Calculate medium properties
    epsilon0_scaled = epsilon0
    mu = mu0 * mu_r
    
    # Convert normalized wavevector to physical wavevector
    # k_phys = k_mpb * (2 * pi / a) radians per meter
    k_phys = k_mpb * (2 * np.pi) / a
    
    # Convert normalized frequency to physical frequency
    # f_phys = f_mpb * (c / a) Hertz
    f_phys = f_mpb * c / a
    
    # Calculate voxel area (since 2D)
    dx, dy = grid_spacing
    voxel_area = dx * dy  # m^2
    
    # Ensure E_mpb and H_mpb are numpy arrays
    E_mpb = np.array(E_mpb)
    H_mpb = np.array(H_mpb)
    
    # Handle permittivity tensor
    epsilon = np.array(epsilon)
    
    # For TE mode: E_mpb shape (Nx, Ny, 2), epsilon shape (Nx, Ny, 2, 2)
    # For TM mode: E_mpb shape (Nx, Ny, 1), epsilon shape (Nx, Ny, 1, 1)
    
    # Compute (E_mpb)^T @ epsilon @ E_mpb for each grid point
    if mode == 'TE':
        # E_mpb has last dimension size 2
        # epsilon has last two dimensions size 2x2
        # Using einsum to compute E^T * epsilon * E for each grid point
        # 'ijk,ijkl,ijl->ij' where:
        # i,j are spatial indices
        # k is E component
        # l is epsilon tensor index
        E_contracted = np.einsum('ijk,ijkl,ijl->ij', E_mpb, epsilon, E_mpb)
    elif mode == 'TM':
        # E_mpb has last dimension size 1
        # epsilon has last two dimensions size 1x1
        # Similar contraction
        E_contracted = np.einsum('ijk,ijkl,ijl->ij', E_mpb, epsilon, E_mpb)
    
    # Compute sum_epsilon_E2 * voxel_area
    sum_epsilon_E2 = np.sum(E_contracted) * voxel_area
    
    # Compute scaling factor for E
    scaling_E = 1.0 / np.sqrt(sum_epsilon_E2)
    
    # Scale the electric field
    E_phys = E_mpb * scaling_E
    
    # Compute sum_H2 * voxel_area
    sum_H2 = np.sum(np.abs(H_mpb)**2) * voxel_area
    
    # Compute scaling factor for H
    scaling_H = 1.0 / np.sqrt(sum_H2)
    
    # Scale the magnetic field
    H_phys = H_mpb * scaling_H
    
    return k_phys, f_phys, E_phys, H_phys
```

```python


import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp2d
plottable_results = []
fig, axs = plt.subplots(len(t_modes), 3, figsize=(15, 5 * len(t_modes)))



for i, mode in enumerate(t_modes):

    e_field, h_field = crystal_2d.convert_mode_fields(mode)
    e_field_z =(e_field[..., 2]).real
    h_field_x =(h_field[..., 0]).real

    # Interpolate the fields for better visualization
    x = np.linspace(0, e_field_z.shape[0], e_field_z.shape[0] * interp)
    y = np.linspace(0, e_field_z.shape[1], e_field_z.shape[1] * interp)
    f_ez = interp2d(np.arange(e_field_z.shape[0]), np.arange(e_field_z.shape[1]), e_field_z, kind='cubic')
    f_hx = interp2d(np.arange(h_field_x.shape[0]), np.arange(h_field_x.shape[1]), h_field_x, kind='cubic')

    e_field_z_interp = f_ez(x, y)
    h_field_x_interp = f_hx(x, y)

    # Plot E-field (z-component)
    axs[i, 0].imshow(e_field_z_interp, extent=[0, e_field_z.shape[0], 0, e_field_z.shape[1]], origin='lower', aspect='auto')
    axs[i, 0].set_title(f'E-field (z) Mode {i}')

    # Plot H-field (x-component)
    axs[i, 1].imshow(h_field_x_interp, extent=[0, h_field_x.shape[0], 0, h_field_x.shape[1]], origin='lower', aspect='auto')
    axs[i, 1].set_title(f'H-field (x) Mode {i}')


    # Calculate the average of E-field (z-component) and H-field (x-component)
    avg_e_field_z = np.mean(e_field_z_interp)
    avg_h_field_x = np.mean(h_field_x_interp)

    Z_eff_z = avg_h_field_x/avg_e_field_z
    
    epsilon_0 = 8.854187817e-12
    mu_0 = 4*np.pi*1e-7
    c0 = 299792458

    eps_eff_z = mode["k_point"][1]/Z_eff_z/epsilon_0/mode["freq"]/c0
    mu_eff_z = -mode["k_point"][1]*Z_eff_z/mu_0/mode["freq"]/c0

    plottable_results.append({
        "Z_eff_z": Z_eff_z,
        "eps_eff_z": eps_eff_z,
        "mu_eff_z": mu_eff_z,
        "freq": mode["freq"],
        "k_point": mode["k_point"],
    })

plt.suptitle("E-field, H-field, and E/H Ratio for Each Mode")
plt.tight_layout()
plt.show()

print(plottable_results)


    
    
```
